```python
def CountFrequency(my_list):
 
    # Creating an empty dictionary
    freq = {}
    for item in my_list:
        if (item in freq):
            freq[item] += 1
        else:
            freq[item] = 1
            
    #del freq['nan']
    for key, value in freq.items():
        print((key, value))

# if __name__ == "__main__":
#     my_list = data['Date of Release'].dt.year
#     CountFrequency(my_list)

diction = CountFrequency(data['Date of Release'].dt.year)
diction
```
